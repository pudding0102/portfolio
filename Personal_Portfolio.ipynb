# **Assignment 4: TCP and UDP Communications**

### Overview
การทดลองนี้ใช้ Packet Tracer เพื่อศึกษาการทำงานของ TCP และ UDP รวมถึงการรับ–ส่งข้อมูล (PDU) ผ่านเครือข่าย และการทำงานของพอร์ต (Port Number) และ Multiplexing

---

### 🔹 Part 1: Network Traffic & Multiplexing
- มีการสร้าง Traffic หลายประเภท เช่น HTTP, FTP, DNS และ Email  
- พบว่าเครือข่ายสามารถส่งข้อมูลหลายโปรโตคอลผ่านสายเดียวกันได้ เรียกว่า **Multiplexing**  
- สีของ PDU แสดงถึงโปรโตคอลที่แตกต่างกัน เช่น TCP และ UDP  

---

### 🔹 Part 2: TCP vs UDP

#### TCP (Reliable)
ใช้ใน:
- HTTP (Port 80)  
- FTP (Port 21)  
- Email: SMTP (25), POP3 (110)  

คุณสมบัติ:
- มี Sequence Number และ Acknowledgement  
- ใช้ 3-way handshake (SYN → SYN-ACK → ACK)  
- รับประกันการส่งข้อมูล (Reliable)

---

#### UDP (Unreliable)
ใช้ใน:
- DNS (Port 53)  

คุณสมบัติ:
- ไม่มี Sequence และ Acknowledgement  
- ส่งข้อมูลเร็ว แต่ไม่รับประกันความถูกต้อง  

---

### 🔹 Key Findings
- TCP เหมาะกับงานที่ต้องการความถูกต้อง เช่น Web, FTP, Email  
- UDP เหมาะกับงานที่ต้องการความเร็ว เช่น DNS  
- Port Number ใช้แยกว่าโปรแกรมไหนเป็นผู้ส่ง/รับข้อมูล

### Overview
การทดลองนี้เป็นการตั้งค่าเครือข่ายพื้นฐาน ประกอบด้วย Router, Switch และ PC เพื่อศึกษาการสื่อสารภายใน LAN รวมถึงการทำงานของ ARP, ICMP และการ Encapsulation

---

### Topology
- อุปกรณ์: R1, S1, PC1, PC2  
- Network: 192.168.1.0/24  
- Default Gateway: 192.168.1.1  

---

### Configuration Summary
- Router (R1): 192.168.1.1  
- Switch (S1): 192.168.1.2  
- PC1: 192.168.1.10  
- PC2: 192.168.1.11  

---

### Connectivity Test
- PC1 สามารถ ping ไป PC2, R1 และ S1 ได้สำเร็จ  
- ARP Table แสดง MAC Address ของอุปกรณ์ในเครือข่าย  
- Switch สามารถเรียนรู้ MAC Address ผ่าน MAC Table  

---

### Protocol Analysis
- ARP: ใช้แปลง IP → MAC Address  
- ICMP: ใช้ในการ ping ทดสอบการเชื่อมต่อ  
- Encapsulation: Data → ICMP → IP → Ethernet  

---

### Troubleshooting
- Subnet Mask ผิด → อาจทำให้สื่อสารไม่ได้  
- Interface shutdown → ทำให้ ping ไม่สำเร็จ  
- Default Gateway ผิด → มีผลเมื่อสื่อสารข้ามเครือข่าย  

---

### Key Learning
- เข้าใจการทำงานของ ARP และ ICMP  
- เห็นกระบวนการ Encapsulation/Decapsulation  
- เข้าใจการทำงานของ Switch และ Router ในเครือข่าย  

---

### Conclusion
เครือข่ายสามารถทำงานได้อย่างถูกต้องเมื่อกำหนดค่า IP, Subnet Mask และ Gateway อย่างเหมาะสม โดยโปรโตคอลต่าง ๆ ทำงานร่วมกันเพื่อให้การสื่อสารสำเร็จ

# **Lab 2: VLAN Design & Router-on-a-Stick**
Screenshot 2026-03-27 230333.png

Link: [Lab 2](https://docs.google.com/document/d/1dMqiBN2k61Y1nLfXnihVHNPONVrl6i2MZSwS6z4F4qY/edit?usp=sharing)

### Overview
การทดลองนี้เป็นการออกแบบเครือข่ายแบบ VLAN เพื่อเพิ่มความปลอดภัยและการจัดการ พร้อมทั้งใช้ Router-on-a-Stick เพื่อให้ VLAN ต่างกันสามารถสื่อสารกันได้

---

### Topology
- อุปกรณ์: R1 (Router), S1 (Switch), PC-A, PC-B, PC-C  
- VLAN:
  - VLAN 10: Users  
  - VLAN 20: Servers  
  - VLAN 99: Management  

---

### Configuration Summary
- ใช้ Subnet /26 (255.255.255.192)  
- Router แบ่ง Subinterface:
  - G0/0.10 → 192.168.10.1  
  - G0/0.20 → 192.168.10.65  
  - G0/0.99 → 192.168.10.129  
- Switch ใช้ trunk (802.1Q) เชื่อม Router  

---

### Key Concepts
- **VLAN**: แยกเครือข่ายเพื่อลด Broadcast และเพิ่มความปลอดภัย  
- **Router-on-a-Stick**: ใช้ Router ตัวเดียว route ระหว่าง VLAN  
- **802.1Q**: ใช้ tag VLAN ใน trunk link  
- **Subnetting (/26)**: แบ่งเครือข่ายออกเป็นหลาย subnet  

---

### Connectivity Test
- PC-A ↔ PC-B (ต่าง VLAN) → สื่อสารได้ผ่าน Router ✅  
- PC ↔ Management VLAN → เข้าถึง Switch ได้ ✅  
- Trunk link ทำงานปกติ (VLAN 10,20,99 ผ่านได้)  

---

### Troubleshooting
- VLAN ผิด → ping ไม่ได้  
- Trunk ไม่อนุญาต VLAN → สื่อสารไม่ได้  
- Gateway ผิด → ติดต่อข้าม VLAN ไม่ได้  
- แก้ไขโดยตรวจ VLAN, trunk และ IP configuration  

---

### Security Design
- VLAN 99 ใช้สำหรับ Management  
- VLAN 999 ใช้ปิดพอร์ตที่ไม่ใช้งาน  
- ลดความเสี่ยงจากการเข้าถึงโดยไม่ได้รับอนุญาต  

---

### Key Learning
- เข้าใจการทำงานของ VLAN และ Inter-VLAN Routing  
- ใช้ Router-on-a-Stick ในการเชื่อมหลาย VLAN  
- เห็นความสำคัญของ Network Security  

---

### Conclusion
VLAN และ Router-on-a-Stick ช่วยให้เครือข่ายมีความเป็นระเบียบ ปลอดภัย และสามารถสื่อสารข้ามเครือข่ายได้อย่างมีประสิทธิภาพ

- Public IP (10.10.x.x) ใช้แทน Internet

### Overview
Lab นี้จำลองระบบเครือข่ายระดับองค์กร โดยมี Internet Edge Router, ISP WAN และ Microservices ที่ทำงานข้ามเครือข่าย  
เพื่อศึกษา NAT, Routing และ Distributed System

---

### Topology
- LAN A: 192.168.10.0/24  
- LAN B: 192.168.20.0/24  
- ISP WAN: 100.10.10.0/30  

อุปกรณ์:
- R1 → Internet Edge (NAT + DHCP WAN)
- R2 → Branch Router
- ServerA / ServerB → Microservices

---

##Network Design
- R1 ทำ NAT (Private → Internet)
- R1 ↔ R2 เชื่อมผ่าน Serial WAN
- ใช้ Static Routing ระหว่าง LAN A และ LAN B
- Internet ออกผ่าน R1

---

### Microservices Deployment
Deploy จาก Week03:

- Upload Service → Port 8000  
- Processing Service → Port 8001  
- AI Service → Port 8002  
- API Gateway → Port 9000  

Server:
- ServerA → 192.168.10.10  
- ServerB → 192.168.20.10  

---

### Testing

#### Cross-Site Communication
curl http://192.168.20.10:8000/health

เส้นทาง:
Client → R1 → ISP → R2 → ServerB

---

#### Full Microservice Workflow
curl -X POST http://192.168.20.10:9000/process-file

Flow:
Upload → Processing → AI → Response

---

#### Internet Test
ping 8.8.8.8

ตรวจสอบ NAT:
show ip nat translations

---

### Key Learning
- NAT ใช้เฉพาะ Internet Edge (R1)
- WAN (Serial) ใช้เชื่อม Branch
- Microservices ทำงานข้ามเครือข่ายได้
- Routing ทำให้ระบบ Distributed เชื่อมต่อกัน

---

### Architecture Insight
- R1 = Internet Gateway  
- R2 = Branch Network  
- ServerA/B = Distributed Services  
- WAN = Inter-site Communication  

---

### Troubleshooting
- Ping ไม่ได้ → ตรวจ Route / WAN  
- NAT ไม่ทำงาน → ตรวจ ACL  
- Service ใช้ไม่ได้ → ตรวจ Port  
- WAN down → ตรวจ Serial Interface  

---

###Conclusion
Lab นี้แสดงให้เห็นการทำงานของระบบจริง  
ที่มีทั้ง Network Infrastructure และ Microservices  
โดยสามารถสื่อสารข้าม WAN และออก Internet ผ่าน NAT ได้

# **New Network**
Screenshot 2026-03-27 230702.png

Spacetime Networking เป็นแนวคิดของเครือข่ายการสื่อสารในอนาคตที่มุ่งลดข้อจำกัดด้านระยะทางและเวลา  
เพื่อให้การรับส่งข้อมูลสามารถเกิดขึ้นได้อย่างรวดเร็ว ต่อเนื่อง และมีประสิทธิภาพสูง แม้ในสภาพแวดล้อมที่ซับซ้อน เช่น อวกาศ ใต้น้ำ หรือโลกเสมือนจริง  

แนวคิดนี้ผสานหลักการด้านเครือข่ายสมัยใหม่เข้ากับทฤษฎีขั้นสูง เช่น การสื่อสารแบบไม่ผูกกับเวลาจริง การคงอยู่ของข้อมูล และการจัดลำดับข้อมูลตามเหตุ–ผล  
เพื่อรองรับการสื่อสารที่มีความไม่แน่นอนและระยะทางไกลระดับจักรวาล  

แม้จะยังมีข้อจำกัดทั้งด้านฟิสิกส์ เทคโนโลยี และความซับซ้อนของระบบ แต่ Spacetime Networking ถูกวางให้เป็นรากฐานของเครือข่ายยุคใหม่  
ที่สามารถพัฒนาไปสู่การสื่อสารข้ามดาว และการเชื่อมต่อของมนุษย์ในอนาคตได้อย่างไร้ขอบเขต
